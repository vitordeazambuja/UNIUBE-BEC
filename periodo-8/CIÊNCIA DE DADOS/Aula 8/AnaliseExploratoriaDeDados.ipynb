```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
data = {
    "id_venda": range(1, 1001),
    "valor_venda": np.random.normal(100, 20, 1000),
    "quantidade": np.random.randint(1, 10, 1000),
    "categoria":np.random.choice(["Eletrônicos", "Roupas", "Alimentos"], 1000),
    "regiao": np.random.choice(["Norte", "Sul", "Leste", "Oeste"], 1000),
    "data_venda": pd.date_range(start="2020-01-01", end="2024-12-31", periods=1000)

}
```

```python
df = pd.DataFrame(data)
```

```python
# Carregando as 5 primeiras linhas do DataFrame

df.head()
```

```python
# Visão Geral dos Dados

df.info()
```

```python
# Estastísticas descritivas

df.describe()
```

```python
# Valores ausentes

df.isnull().sum()
```

```python
# Análise de Distribuiçãoo do valor das vendas

plt.figure(figsize=(10, 6))
sns.histplot(df["valor_venda"], bins=20, kde=True)
plt.title("Distribuição do Valor das Vendas")
plt.xlabel("Valor da Venda")
plt.ylabel("Frequência")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="categoria", y="valor_venda", data=df)
plt.title("Boxplot do Valor das Vendas por Categoria")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="regiao", y="valor_venda", data=df)
plt.title("Boxplot do Valor das Vendas por Região")
```
