#### Conversão de variáveis categoricas

O que são variáveis categóricas?

São aquelas variáveis que representam categorias, como: "cor", "sexo" ou categoria de produtos



1.   Ecoding de Variáveis categoricas:
    * **Label encoding:** Atribui um valor númerico único para cada categoria. Pode ser útil quando as categorias têm ordem intrínseca.
    * **One Hot encoding** Cria uma nova variável binária para cada categoria. Usando principalmente quando não há ordem entre as categorias




```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.DataFrame(
    {
        'cor': [ 'Vermelho', 'Azul', 'Verde', 'Azul', 'Vermelho'],
        'Sexo': ['Masculino', 'Feminino', 'Feminino', 'Masculino', 'Feminino']

    }
)
```

```python
df
```

```python
#Label Encoding

label_enconder = LabelEncoder()
df['cor_label'] = label_enconder.fit_transform(df['cor'])
df['sexo_label'] = label_enconder.fit_transform(df['Sexo'])
```

```python
df
```

```python
one_hot_encoder = OneHotEncoder()
df_onehot = one_hot_encoder.fit_transform(df[['cor']]).toarray()
```

```python
df_onehot
```

### EDA Análise Exploratpória de Dados

A análise exploratória de dados (EDA) é uma etapa crucial para entender o comportamento dos dados e gerar insigths. Aqui, vamos explotat os dados por meio de vizualizações



1.   **Visualizações Básica**
    * **Histograma:** Para ver a disttuição de uma variável númerica.
    * **Boxplot:** Para verificar outilers e a distruibuição de varíavel númerica.
    * **Gráfico de barras:** Para mostrar a distribuição de variáveis categoricas.




```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(
    {
        'Idade': [25, 32, 45, 23, 36],
        'Salario': [2500, 3500, 4500, 3000, 4000],
        'Sexo': ['Masculino', 'Feminino', 'Feminino', 'Masculino', 'Feminino']

    }
)
```

```python
df
```

```python
#Histograma
plt.figure(figsize=(10, 6))
sns.histplot(df['Idade'], kde=True, color='blue')
plt.title('Histograma de Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
#Histograma
plt.figure(figsize=(10, 6))
sns.histplot(df['Salario'], kde=True, color='blue')
plt.title('Histograma de Salario')
plt.xlabel('Salário')
plt.ylabel('Frequência')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sexo', y='Salario', data=df)
plt.title('Boxplot de distribuição de salario por sexo')
plt.show()
```
