```python
import numpy as np
from scipy.spatial.distance import euclidean
import matplotlib.pyplot as plt


def menor_distancia (centroides_distancia):
  return np.argmin(centroides_distancia)

def agrupar_por_labels(data, labels):
    grupos = {}
    for ponto, label in zip(data, labels):
        if label not in grupos:
            grupos[int(label)] = []
        grupos[int(label)].append(ponto.tolist())

    #print([[label, pontos] for label, pontos in grupos.items()])
    return [[label, pontos] for label, pontos in grupos.items()]

def calcular_distancias(data, centroides):
  lDados  = len(data)
  lCent   = len(centroides)
  resultado = np.zeros(lDados)

  for i in range(lDados):
        array_distancia_centroide = np.zeros(lCent)
        for j in range(lCent):
            dist = euclidean(data[i], centroides[j] )
            print (f"Distância C{j} [{data[i]}{centroides[j]} = {dist}")
            array_distancia_centroide[j] = dist
        resultado[i] = menor_distancia(array_distancia_centroide)

  return resultado

def calcular_media_centroides(grupos):
  centroides_dict = {label: np.mean(pontos, axis=0) for label, pontos in grupos}
  centroides = np.array([centroides_dict[key] for key in sorted(centroides_dict.keys())])
  return centroides

def calcular_coeficiente_dunn(data, labels):
    unique_labels = np.unique(labels)
    clusters = {label: data[labels == label] for label in unique_labels}

    # Calcular a maior distância intra-cluster
    max_intra_cluster_dist = 0
    for pontos in clusters.values():
        if len(pontos) > 1:
            intra_dists = [euclidean(p1, p2) for i, p1 in enumerate(pontos) for p2 in pontos[i+1:]]
            max_intra_cluster_dist = max(max_intra_cluster_dist, max(intra_dists, default=0))

    # Calcular a menor distância inter-cluster
    min_inter_cluster_dist = float('inf')
    for i, label_i in enumerate(unique_labels):
        for j, label_j in enumerate(unique_labels):
            if label_i != label_j:
                inter_dists = [euclidean(p1, p2) for p1 in clusters[label_i] for p2 in clusters[label_j]]
                min_inter_cluster_dist = min(min_inter_cluster_dist, min(inter_dists, default=float('inf')))

    # Evita divisão por zero
    if max_intra_cluster_dist == 0:
        return float('inf')

    return min_inter_cluster_dist / max_intra_cluster_dist

data = np.array([[2.50,1.86],
[2.65,3.52],
[1.77,1.77],
[3.58,2.77],
[1.53,2.54],
[1.54,1.53],
[2.24,0.09],
[0.28,1.44],
[0.99,2.31],
[1.09,0.59]])

centroides = np.array([
    [1.53,2.54],
    [1.09,0.59],
    [2.23,2.63]
])

labels = calcular_distancias(data,centroides)

grupos = agrupar_por_labels(data,labels)

print(f"Grupos 1 -{grupos}\n")

centroides2 = calcular_media_centroides(grupos)

print(f"Medias -{centroides2}\n")

print(f"Coeficiente Dunn - 1{calcular_coeficiente_dunn(data,labels)}")

labels = calcular_distancias(data,centroides2)

grupos = agrupar_por_labels(data,labels)

print(f"Grupos 2 -{grupos}\n")

print(f"Coeficiente Dunn - 2{calcular_coeficiente_dunn(data,labels)}")


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = np.array([
    [2.50, 1.86],
    [2.65, 3.52],
    [1.77, 1.77],
    [3.58, 2.77],
    [1.53, 2.54],
    [1.54, 1.53],
    [2.24, 0.09],
    [0.28, 1.44],
    [0.99, 2.31],
    [1.09, 0.59]
])

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

# Treinar modelo
kmeans.fit(data)

# Obter rótulos e centroides
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("Rótulos dos clusters:", labels)
print("Coordenadas dos centroides:", centroids)

# Visualização dos clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroides')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('K-Means Clustering')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Carregar a imagem
image = cv2.imread('image.jpg')  # Substitua pelo caminho da sua imagem
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Converter de BGR para RGB

# Redimensionar para um vetor 2D (cada pixel como um ponto)
pixels = image.reshape((-1, 3))  # Formato (altura * largura, 3)

# Definir número de clusters (cores)
K = 3  # Número de segmentos
kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)

# Ajustar K-Means aos dados de pixel
kmeans.fit(pixels)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Criar imagem segmentada substituindo cada pixel pelo centroide correspondente
segmented_image = centroids[labels].reshape(image.shape).astype(np.uint8)

# Exibir imagem original e segmentada
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(image)
ax[0].set_title('Imagem Original')
ax[0].axis('off')

ax[1].imshow(segmented_image)
ax[1].set_title(f'Imagem Segmentada (K={K})')
ax[1].axis('off')

plt.show()
```
